```python
import pandas as pd
```

### The first dataset was found on Google Dataset Search (https://www.johnsnowlabs.com/). Information of 2019.

```python
url="https://raw.githubusercontent.com/danieltp92/datasets/main/data.csv"
data = pd.read_csv(url)
data
```

### Reviewing the information from the dataset, we can see that it has 18 variables and a total of 352 records

```python
data.info()
```

## Columns relevant for study

#### The country and number of countries would be relevant to see a patern in the relationship between tech companies and how impact in mental health

```python
list_country = data['Country'].unique()
list_country
```

#### We make a count of how many countries are involved in the study

```python
def num_elem(list):
    count = 0
    for element in list:
        count += 1
    return count

print("Number of countries: ", num_elem(list_country))
```

#### We check the size of the companies and found that some records need to be cleaned due the fact that the system take some as dates.

```python
print(data['Number_of_Employees_in_Organization'].unique())
```

#### We check how many companies are related to technology

```python
tech_df = pd.DataFrame(data, columns=['Is_Employer_Tech_Company'])
dups = tech_df.pivot_table(columns=['Is_Employer_Tech_Company'], aggfunc='size')
print(dups)
```

#### The age is relevant to figure if it has any relationship with mental health. Also we can make groups for a data model.

```python
tech_df = pd.DataFrame(data, columns=['Age'])
dups = tech_df.pivot_table(columns=['Age'], aggfunc='size')
print(dups)
```
